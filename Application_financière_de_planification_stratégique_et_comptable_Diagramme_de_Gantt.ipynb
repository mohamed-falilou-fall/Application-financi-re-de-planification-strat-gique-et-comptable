```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# -----------------------------------
# Donnes du projet
# -----------------------------------
data = [
    ("Phase 1 - Cadrage du projet", "Cadrage", "Mohamed Falilou Fall", "2025-12-24", "2025-12-24"),
    ("Phase 1 - Architecture & Stack", "Cadrage", "Mohamed Falilou Fall", "2025-12-24", "2025-12-24"),
    ("Phase 2 - Hypothses dactivit", "Modlisation financire", "Mohamed Falilou Fall", "2025-12-25", "2025-12-25"),
    ("Phase 3 - Compte de rsultat", "Modlisation financire", "Mohamed Falilou Fall", "2025-12-26", "2025-12-26"),
    ("Phase 4 - Seuil de rentabilit", "Analyse financire", "Mohamed Falilou Fall", "2025-12-27", "2025-12-27"),
    ("Phase 5 - BFR", "Analyse financire", "Mohamed Falilou Fall", "2025-12-27", "2025-12-27"),
    ("Phase 6 - Plan de financement", "Analyse financire", "Mohamed Falilou Fall", "2025-12-28", "2025-12-28"),
    ("Phase 7 - Plan de trsorerie", "Analyse financire", "Mohamed Falilou Fall", "2025-12-28", "2025-12-28"),
    ("Phase 8 - KPI & Ratios", "Analyse financire", "Mohamed Falilou Fall", "2025-12-29", "2025-12-29"),
    ("Phase 9 - Scnarios & Sensibilit", "Analyse financire", "Mohamed Falilou Fall", "2025-12-29", "2025-12-29"),
    ("Phase 10 - Exports Excel & PDF", "Restitution", "Mohamed Falilou Fall", "2025-12-30", "2025-12-30"),
    ("Phase 11 - UX, Scurit & Tests", "Restitution", "Mohamed Falilou Fall", "2025-12-30", "2025-12-30"),
    ("Phase 12 - Documentation & Dploiement", "Restitution", "Mohamed Falilou Fall", "2025-12-30", "2025-12-30"),
]

df = pd.DataFrame(data, columns=["Phase", "Type", "Responsable", "Start", "Finish"])
df["Start"] = pd.to_datetime(df["Start"])
df["Finish"] = pd.to_datetime(df["Finish"])

# -----------------------------------
# Couleurs par type de tche
# -----------------------------------
colors = {
    "Cadrage": "#4C72B0",
    "Modlisation financire": "#55A868",
    "Analyse financire": "#C44E52",
    "Restitution": "#8172B3"
}

# -----------------------------------
# Cration du diagramme de Gantt
# -----------------------------------
fig, ax = plt.subplots(figsize=(16, 9))

for i, task in df.iterrows():
    ax.barh(
        task["Phase"],
        (task["Finish"] - task["Start"]).days + 1,
        left=task["Start"],
        color=colors[task["Type"]],
        edgecolor="black"
    )

# Mise en forme des dates
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%d %b"))

# Titre et labels
ax.set_title(
    "Diagramme de Gantt  Application financire de planification stratgique et comptable\n"
    "Ressource : Mohamed Falilou Fall (100%)",
    fontsize=16,
    weight="bold"
)
ax.set_xlabel("Dates (Dcembre 2025)")
ax.set_ylabel("Phases du projet")

# Lgende personnalise
legend_handles = [
    plt.Rectangle((0, 0), 1, 1, color=color)
    for color in colors.values()
]
ax.legend(legend_handles, colors.keys(), title="Type de tche")

# Ordre lecture
ax.invert_yaxis()

# Grille
ax.grid(axis="x", linestyle="--", alpha=0.4)

# -----------------------------------
# Export
# -----------------------------------
output_file = "diagramme_gantt_streamlit_financier_matplotlib.png"
plt.tight_layout()
plt.savefig(output_file, dpi=300)
plt.show()

print(f" Diagramme export en PNG : {output_file}")
```
